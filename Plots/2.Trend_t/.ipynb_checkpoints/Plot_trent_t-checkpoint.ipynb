```python
import numpy as np
import h5py
from tqdm import tqdm
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# initialize
njob = 36
tpulse_list = np.arange(0,48)
Natom = 1000
dsamp = 10 * 1e-9 		# sample thickness (m)
Tmodels = ['Square','Square_mono','Gaus','Gaus_mono','SASE','SASE_mono']
colors = ['sandybrown','sandybrown','grey','grey','mediumseagreen','mediumseagreen']

dirname = 'Plots/2.Trend_t/'

contrast = np.zeros((len(Tmodels), tpulse_list.size))
error = np.zeros_like(contrast)
```

```python
# load data
for i, Tmodel in enumerate(Tmodels):
    print('loading', Tmodel)
    dircase = '{}/'.format(Tmodel)
    cont_holder = []
    for j, tpulse in tqdm(enumerate(tpulse_list)):
        for numjob in range(njob):
            with h5py.File(dircase+'{}atom_{}fs_{}_{}nm.h5'.format(Natom,float(tpulse),Tmodel,round(dsamp*1e9,2)), 'r') as f:
                try: cont_holder.append(f['{}/contrast'.format(numjob)][0])
                except:
                    print('   ',Tmodel, tpulse, numjob, 'missing')
        contrast[i, j] = np.mean(cont_holder)
        error[i,j] = np.std(cont_holder)
```

```python
# measurement
PL95 = np.array([7, 11, 14, 35]);
PL95E = np.array([3,5,7,15]);
C95 = np.array([0.337, 0.312, 0.310, 0.305])-0.270;
C95E = np.array([0.01, 0.005, 0.006, 0.01]);

PL72 = np.array([15,40])
PL72E = np.array([6, 15])
C72 = np.array([0.279,0.265]) - 0.260
C72E = np.array([0.006,0.002])

# # ax2.axes.set_yticks([0,0.04,0.08])
# #ax2.axes.set_ylim(ax1.axes.get_ylim())
# #ax2.axes.get_yaxis().set_visible(False)
# ax2.legend(frameon=False)
# ax2.axes.set_ylabel('contrast')
# ax2.axes.set_xlabel('pulse duration (fs)')
# plt.tight_layout()
```

```python
# plot simulation
fig, (ax1, ax2) = plt.subplots(2, sharex=True, figsize=(5,12))
for i, Tmodel in enumerate(Tmodels):
    if not Tmodel.endswith('mono'):
        ax1.errorbar(tpulse_list, contrast[i], yerr=error[i]/np.sqrt(njob),
                     label=Tmodel, color=colors[i])
    else:
        ax2.errorbar(tpulse_list, contrast[i], yerr=error[i]/np.sqrt(njob),
                     label=Tmodel, color=colors[i])

# plot measurement
ax2.errorbar(PL95, C95, C95E, PL95E, fmt='.', marker='s', color='r', markersize=8, label='9.5 keV')
ax2.errorbar(PL72, C72, C72, PL72E, fmt='.', marker='s', color='b', markersize=8, label='7.2 keV')

# set plot style      
ax1.legend(frameon=False); ax2.legend(frameon=False)
ax1.set_xlabel('pulse duration (fs)'); ax2.set_xlabel('pulse duration (fs)')
ax1.set_ylabel('contrast'); ax2.set_ylabel('contrast')


```
