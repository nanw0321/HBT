```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load
njobs = 500
contrasts = np.zeros(njobs)
tpulses = np.zeros(njobs)
nl_ts = np.zeros(njobs)

with h5py.File('Gaus/1000atom_Gaus_10.0nm.h5','r') as f:
    for njob in range(njobs):
        contrasts[njob] = f['{}/contrast'.format(njob)][0]
        nl_ts[njob] = f['{}/others'.format(njob)][1]
        tpulses[njob] = 1/nl_ts[njob] / (2*np.sqrt(np.pi))*2.36
```

```python
plt.figure()
plt.scatter(contrasts, nl_ts, c=np.log10(tpulses), cmap='jet', marker='.',alpha=0.8)
plt.xscale('log'); plt.yscale('log')
plt.xlabel('contrast'); plt.ylabel(r'$S_S$')
```
