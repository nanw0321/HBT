```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
# case by case debugging
njob = 0

with h5py.File('Gaussian_non-linear_N=1000_10.0nm/Gaussian_r=1_sig=1-100um.h5','r') as f:
    img = f['{}/pattern'.format(njob)][:]
    img_sub = f['{}/pattern_sub'.format(njob)][:]
    contrast=f['{}/contrast'.format(njob)][0]
    others = f['{}/others'.format(njob)][:]
    spotsig=others[0]
    nl_s=others[2]

plt.figure(figsize=(12,5))
plt.subplot(1,2,1); plt.imshow(img/img.mean(), cmap='jet',vmin=0, vmax=3)
plt.subplot(1,2,2); plt.imshow(img_sub/img_sub.mean(), cmap='jet', vmin=0, vmax=3)

print('job:{}, beam size:{}um, contrast:{}'.format(njob, round(spotsig*1e6,2), round(contrast,3)))
```

```python
# load
njobs = 180
contrasts = np.zeros(njobs)
spotsigs = np.zeros(njobs)
nl_ss = np.zeros(njobs)

with h5py.File('Gaussian_non-linear_N=1000_10.0nm/Gaussian_r=1_sig=1-100um.h5','r') as f:
    for njob in range(njobs):
        contrasts[njob]=f['{}/contrast'.format(njob)][0]
        others = f['{}/others'.format(njob)][:]
        spotsigs[njob]=others[0]
        nl_ss[njob]=others[2]
```

```python
plt.figure()
plt.scatter(contrasts, nl_ss, c=spotsigs*1e6, cmap='jet', marker='.',alpha=1.0)
plt.xscale('log'); plt.yscale('log')
plt.xlabel('contrast'); plt.ylabel(r'$S_S$')
```
